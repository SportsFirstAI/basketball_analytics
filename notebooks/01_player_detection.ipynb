```python
import cv2
import supervision as sv
from rfdetr import RFDETRSmall

# 1. Set your paths here!
# Replace with the exact path you found from the search script
MODEL_WEIGHTS_PATH = "/content/output/checkpoint_best_ema.pth"

# Replace with the name of the video you uploaded to test
SOURCE_VIDEO_PATH = "/stephcurryvideo.mp4"
TARGET_VIDEO_PATH = "output_video.mp4"

# 2. Load your custom-trained RF-DETR model
print("Loading model...")
# Fix: Explicitly set num_classes to match the checkpoint's class count
model = RFDETRSmall(pretrain_weights=MODEL_WEIGHTS_PATH, num_classes=10)
# Optional: Optimize the model for faster inference
model.optimize_for_inference()
print("Model loaded successfully!")

# 3. Extract video metadata and create a frame generator
video_info = sv.VideoInfo.from_video_path(SOURCE_VIDEO_PATH)
generator = sv.get_video_frames_generator(SOURCE_VIDEO_PATH)

# 4. Initialize supervision annotators for drawing bounding boxes and text
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# 5. Loop through the video frames, run inference, and save the results
print("Processing video...")
with sv.VideoSink(target_path=TARGET_VIDEO_PATH, video_info=video_info) as sink:
    for frame in generator:
        # Run inference (converting BGR to RGB format for RF-DETR)
        # You can lower the threshold (e.g., 0.3) if it's missing players,
        # or raise it (e.g., 0.7) if it's detecting too many false objects.
        detections = model.predict(frame[:, :, ::-1].copy(), threshold=0.5) # Added .copy() here

        # Create labels dynamically using the class names from your trained model
        labels = [
            f"{model.class_names[class_id]} {confidence:.2f}"
            for class_id, confidence in zip(detections.class_id, detections.confidence)
        ]

        # Annotate the frame with boxes and labels
        annotated_frame = frame.copy()
        annotated_frame = box_annotator.annotate(scene=annotated_frame, detections=detections)
        annotated_frame = label_annotator.annotate(scene=annotated_frame, detections=detections, labels=labels)

        # Save the frame to the new video
        sink.write_frame(frame=annotated_frame)

print(f"вң… Inference complete! Annotated video saved to: {TARGET_VIDEO_PATH}")
```
